```python
import pandas as pd
```

```python
star_wars = pd.read_csv("star_wars.csv")
```

```python
star_wars.head()
```

```python
star_wars = star_wars[1:]
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map({"Yes":True, "No":False})
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts(dropna = False)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map({"Yes":True, "No":False})
```

```python
star_wars
```

```python
import numpy as np
```

```python
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}
```

```python
star_wars.iloc[:,3:9]  = star_wars.iloc[:,3:9].replace(movie_mapping)
```

```python
star_wars.head()
```

```python
star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"}, inplace = True)
```

```python
star_wars
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1",
                          "Unnamed: 10":"ranking_2","Unnamed: 11":"ranking_3","Unnamed: 12":"ranking_4","Unnamed: 13":"ranking_5",
                          "Unnamed: 14":"ranking_6"}, inplace = True)
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
star_wars[star_wars.columns[9:15]].mean().plot.bar()
```

So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each movie. As I suspected, it looks like the "original" movies are rated much more highly than the newer ones.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
star_wars[star_wars.columns[3:9]].sum().plot.bar()
```

It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
males[males.columns[9:15]].mean().plot.bar()
plt.show()

females[females.columns[9:15]].mean().plot.bar()
plt.show()
```

```python
males[males.columns[3:9]].sum().plot.bar()
plt.show()

females[males.columns[3:9]].sum().plot.bar()
plt.show()
```
